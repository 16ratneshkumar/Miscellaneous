```python
# Import Necessary Libaray
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Data Loading and Preprocessing
df = pd.read_csv('./Salary_dataset.csv')
x = df.loc[:,'YearsExperience'].values.reshape(-1, 1)
y = df.iloc[:,-1].values

# Data Splitting
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)

# Select Model and Training
model = LinearRegression()
model.fit(X_train,y_train)

# Model Testing
y_pred = model.predict(X_test)

# Model Evalution
print("The value of the coefficient of determinaton(R^2) is :: ",round(r2_score(y_pred,y_test),2))

# Prediction
YearsExperience = np.array(float(input("Enter year of experience :: "))).reshape(-1, 1)
User_pred = model.predict(YearsExperience)
print("After ",YearsExperience[0][0]," years of experience your salary is:: ",round(User_pred[0],2))

# Visualization
plt.scatter(X_train,y_train,color="red",marker='s',s=100,alpha=0.5,label="Training Data")
plt.scatter(X_test,y_pred,color="green",marker='^',s=100,alpha=0.5,label="Test Data Prediction")
plt.scatter(YearsExperience,User_pred,marker='*',s=100,alpha=1,label="User Input Prediction")
plt.xlabel("Years Of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
